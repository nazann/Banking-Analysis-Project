```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bins=[0,100000,300000, float("inf")]
labels=['Low','Medium','High']
df['Income Band']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

## First Insights ##

## univariate analisys

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
cat_cols=df[["Amount of Credit Cards","Properties Owned","BRId" ,"GenderId","IAId","Income Band","Nationality",
             "Risk Weighting","Occupation","Fee Structure","Loyalty Classification"]]

for i,col in enumerate(cat_cols):
    plt.figure(i)
    sns.countplot(data=df,x=col)
```

## Bivariate A.

```python
for i,col in enumerate(cat_cols):
    plt.figure(i)
    sns.countplot(data=df,x=col,hue='Nationality')
```

```python
for i,col in enumerate(cat_cols):
    plt.figure(i)
    sns.countplot(data=df,x=col,hue='Risk Weighting')
```

## Numerical A.

```python
num_cols=df[["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits",
             "Checking Accounts","Saving Accounts","Business Lending"]]
```

```python
plt.figure(figsize=(15,10))
for i,col in enumerate(num_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

plt.show()
```

## CORR heatmaps

```python
cols=["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits",
             "Checking Accounts","Saving Accounts","Business Lending","Foreign Currency Account"]
corr_matrix=df[cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(corr_matrix,annot=True,cmap="crest")
plt.show()
```

Strong Correlation Between Bank Deposits and Checking Accounts (r â‰ˆ 0.85) 
-This indicates that checking account balances contribute significantly to the total bank deposits. Customers with higher checking account balances tend to hold larger overall deposits, reflecting the central role of checking accounts in daily financial activities.

High Correlation Between Bank Deposits and Saving Accounts
-Savings accounts also make up a substantial portion of total deposits. The positive relationship suggests customers maintain substantial funds in savings alongside their checking accounts, reinforcing the idea that total bank deposits aggregate these key account types.

High Correlation Between Bank Deposits and Business Lending
-The link between deposits and lending highlights that business clients with larger deposit balances often have higher loan amounts. This may reflect the scale of business operations or the depth of banking relationships where deposit and credit products are bundled.
